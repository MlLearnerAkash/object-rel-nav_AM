```python
import pickle


# Load all objects from the pickle file (supports single or multiple objects)
all_data = []
with open("/data/ws/object-rel-nav/data/hm3d_iin_val/4ok3usBNeis_0000000_chair_8_/nodes_gt_topometric.pickle", "rb") as f:
    while True:
        try:
            all_data.append(pickle.load(f))
        except EOFError:
            break

print(f"Loaded {len(all_data)} object(s).")
data = all_data[0] if len(all_data) == 1 else all_data

# If the loaded object is a NetworkX graph, extract full content
if hasattr(data, "nodes") and hasattr(data, "edges"):
    graph_data = {
        "graph_attrs": dict(data.graph),
        "nodes": {n: attrs for n, attrs in data.nodes(data=True)},
        "edges": [(u, v, attrs) for u, v, attrs in data.edges(data=True)],
    }

    print("Graph attrs:", graph_data["graph_attrs"])
    print("Num nodes:", len(graph_data["nodes"]))
    print("Num edges:", len(graph_data["edges"]))
```
